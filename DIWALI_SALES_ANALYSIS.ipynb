# EDA & Data Visualization

```python
df.columns
```

```python
a=sns.countplot(data=df,x='Gender')
a.bar_label(a.containers[0])
plt.title("Gender-Distribution")
plt.show()
```

```python
#conclusion=from above chart we can say the sales done by female is more than male
```

```python
x=df.groupby("Gender").agg({"Orders":"sum","Amount":"sum"})
```

```python
x
```

```python
sns.barplot(data=x,x='Orders',y='Amount',palette='Greens_r')
plt.show()
```

```python
#conclusion=orders placed by female is greater
```

```python
z=sns.countplot(data=df,x='Age_Group')
z.bar_label(z.containers[0])
plt.title("Age_Group_Distribution")
plt.show()
```

```python
#we conclude that age between 26-35 was done the sales good
```

```python
s=df.groupby("State").agg({'Orders':'sum'})
s
```

```python
q=sns.heatmap(s,annot=True,fmt='')
plt.title("State wise Orders")

plt.show()
```

```python
#we can conclude from above grouby we find total orders by each state
```

```python
d=df.groupby("Occupation").agg({'Orders':'sum','Amount':'sum'})
d
```

```python
p=sns.heatmap(d,annot=True,fmt='',cmap='pink',linewidth=2)
plt.title("Sector wish orders with amounts")
plt.show()
```

```python
#from above we can say peoples in which sector done more shopping
```

```python
sns.set({'figure.figsize':(25,5)})
f=sns.countplot(data=df,x='Occupation')
f.bar_label(f.containers[0])
plt.show()
```

```python
#conclude from above itsector aviation and healthcare and banking occupations people was done more shopping
```

```python
sns.set({'figure.figsize':(25,5)})
f=sns.countplot(data=df,x='Product_Category')
f.bar_label(f.containers[0])
plt.show()
```

```python
#conclude from above more products that was sold in categoy food,clothing,electronics
```

```python
sns.set({'figure.figsize':(25,5)})
l=sns.countplot(data=df,x='State')
l.bar_label(l.containers[0])
plt.show()
```

```python
#conclude from above more shopping was done in Up,maharastra,karnataka
```

```python
e=df.groupby('Marital_Status').agg({'Orders':'count','Amount':'sum'})
e
```

```python
f=sns.countplot(data=df,x='Marital_Status')
f.bar_label(f.containers[0])
plt.show()
```

```python
#conclude from above mostly married peoples shopping more
```
